```python
from sklearn.preprocessing import MinMaxScaler
```

```python
wqi_params = data[['D.O. (mg/l)  : Mean  : > 4 mg/l',
                   'pH  : Mean  : 6.5-8.5',
                   'B.O.D. (mg/l)  : Mean  : < 3 mg/l',
                   'FECAL COLIFORM (MPN/100ml)  : Mean  : < 2500 MPN/100ml',
                   'TOTAL COLIFORM (MPN/100ml)  : Mean  : < 5000 MPN/100ml',
                   'CONDUCTIVITY (µmhos/cm)  : Mean',
                   'NITRATE- N+ NITRITE-N (mg/l)  : Mean']]

scaler = MinMaxScaler(feature_range=(0, 100))

wqi_normalized = pd.DataFrame(scaler.fit_transform(wqi_params), columns=wqi_params.columns)

print(wqi_normalized.head())
```

```python
wqi_normalized.head()
```

```python

weights = {
    'D.O. (mg/l)  : Mean  : > 4 mg/l': 0.25,
    'pH  : Mean  : 6.5-8.5': 0.15,
    'B.O.D. (mg/l)  : Mean  : < 3 mg/l': 0.20,
    'FECAL COLIFORM (MPN/100ml)  : Mean  : < 2500 MPN/100ml': 0.20,
    'TOTAL COLIFORM (MPN/100ml)  : Mean  : < 5000 MPN/100ml': 0.10,
    'CONDUCTIVITY (µmhos/cm)  : Mean': 0.05,
    'NITRATE- N+ NITRITE-N (mg/l)  : Mean': 0.05
}

wqi_normalized['WQI'] = (
    wqi_normalized['D.O. (mg/l)  : Mean  : > 4 mg/l'] * weights['D.O. (mg/l)  : Mean  : > 4 mg/l'] +
    wqi_normalized['pH  : Mean  : 6.5-8.5'] * weights['pH  : Mean  : 6.5-8.5'] +
    wqi_normalized['B.O.D. (mg/l)  : Mean  : < 3 mg/l'] * weights['B.O.D. (mg/l)  : Mean  : < 3 mg/l'] +
    wqi_normalized['FECAL COLIFORM (MPN/100ml)  : Mean  : < 2500 MPN/100ml'] * weights['FECAL COLIFORM (MPN/100ml)  : Mean  : < 2500 MPN/100ml'] +
    wqi_normalized['TOTAL COLIFORM (MPN/100ml)  : Mean  : < 5000 MPN/100ml'] * weights['TOTAL COLIFORM (MPN/100ml)  : Mean  : < 5000 MPN/100ml'] +
    wqi_normalized['CONDUCTIVITY (µmhos/cm)  : Mean'] * weights['CONDUCTIVITY (µmhos/cm)  : Mean'] +
    wqi_normalized['NITRATE- N+ NITRITE-N (mg/l)  : Mean'] * weights['NITRATE- N+ NITRITE-N (mg/l)  : Mean']
)


wqi_normalized[['WQI']].head()

wqi_normalized.to_excel('WQI_results.xlsx', index=False)
```

```python
wqi_normalized.head()
```

```python
wqi_normalized["WQI"].min()
```

```python
wqi_max = 51.75414862914861
wqi_min = 20.026139107895595


wqi_normalized['WQI_scaled'] = (wqi_normalized['WQI'] - wqi_min) / (wqi_max - wqi_min) * 100


wqi_normalized[['WQI', 'WQI_scaled']].head()
```

```python
def categorize_wqi(wqi):
    if wqi <= 25:
        return 'Very Poor'
    elif 25 < wqi <= 50:
        return 'Poor'
    elif 50 < wqi <= 75:
        return 'Moderate'
    else:
        return 'Good'

wqi_normalized['WQI_category'] = wqi_normalized['WQI_scaled'].apply(categorize_wqi)


wqi_normalized.to_excel('WQI_categorized.xlsx', index=False)
```

```python
wqi_normalized.drop(["WQI","WQI_scaled"],inplace=True,axis=1)
```

```python
wqi_normalized.head()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

wqi_normalized['WQI_category'] = label_encoder.fit_transform(wqi_normalized['WQI_category'])
```

```python
wqi_normalized.head()
```

```python
wqi_normalized.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = wqi_normalized.drop("WQI_category",axis=1)
Y = wqi_normalized["WQI_category"]
```

```python
X.head()
```

```python
Y.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
X_train.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_absolute_error

param_grid = {
    'n_estimators': [100,150, 200,300],
    'max_features': ['log2', 'sqrt',5,11],
    'max_depth': [None,1, 10,20],
    'min_samples_split': [2, 5,6,7,8],
    'min_samples_leaf': [1,2,3,4]
}


rf_random = RandomizedSearchCV(estimator=model, param_distributions=param_grid, n_iter=50, cv=3, random_state=42, n_jobs=-1)

rf_random.fit(X_train, Y_train)
```

```python
print(f'Best Parameters: {rf_random.best_params_}')
y_pred_tuned = rf_random.predict(X_test)
mae_tuned = mean_absolute_error(Y_test, y_pred_tuned)
print(f'Tuned Model Mean Absolute Error: {mae_tuned}')
```

```python
from sklearn.metrics import r2_score
r2_score(Y_test,y_pred_tuned)
```
